```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pwm_to_lift(pwm):
    a = 7.5*10**-7
    b = -0.001325
    c = 0.55
    return (a* pwm**2 + b*pwm + c)*9.81

pwm = np.arange(1100, 2100, 1)
lift = pwm_to_lift(pwm)


plt.figure()
plt.plot(pwm, lift)
plt.xlabel('PWM')
plt.ylabel('Lift [N]')
plt.title('Lift vs PWM')

plt.figure()
plt.plot(pwm, np.sqrt(lift))
plt.xlabel('PWM')
plt.ylabel('sqrt(Lift) [N]')
plt.title('sqrt(Lift) vs PWM')


plt.figure()
plt.plot(np.sqrt(lift), pwm)
plt.xlabel('sqrt(Lift) [N]')
plt.ylabel('PWM')
plt.title('PWM vs sqrt(Lift)')
```

```python
def lift_to_pwm(lift):
    a = 7.5*10**-7
    b = -0.001325
    c = 0.55
    return (-b + np.sqrt(b**2 - 4*a*(c-lift/9.81)))/(2*a)

lift_req = np.arange(0, 10, 0.1)
pwm_req_exact = lift_to_pwm(lift_req)

# plt.figure()
# plt.plot(lift, pwm)


def lif_to_pwm_simple(lift):
    lift_sqrt = np.sqrt(lift)   
    return 364.5*lift_sqrt + 945

pwm_req_simple = lif_to_pwm_simple(lift_req)

plt.figure()
plt.plot(lift_req, pwm_req_exact, 'r')
plt.plot(lift_req, pwm_req_simple, 'b--')
plt.xlabel('Lift [N]')
plt.ylabel('PWM')
plt.title('PWM vs Lift')
plt.legend(['Exact', 'Simple'])

```
