# Filters

```python
%matplotlib widget
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ts = np.linspace(0, 10, 1000)
y = np.sin(ts) + np.random.normal(0, 0.1, ts.shape)

plt.figure()
plt.plot(ts, y)
plt.show()
```

# Average Window

```python
# average over N windows
N = 100
y_avg = np.zeros(y.shape)
cnt = 0
windowSum = 0
for i in range(0, len(y)):
    if cnt < N:
        windowSum = windowSum + y[i]
        cnt = cnt + 1
    if cnt == N:
        y_avg[i-N:i] = windowSum / N
        windowSum = 0
        cnt = 0
    
    


plt.figure()
plt.plot(ts, y)
plt.plot(ts, y_avg)
plt.show()
```

# Low Pass Filter

```python
# low pass filter

# y[n] = a * y[n-1] + (1-a) * x[n]
a = 0.9
y_lp = np.zeros(y.shape)
for i in range(1, len(y)):
    y_lp[i] = a * y_lp[i-1] + (1-a) * y[i]

plt.figure()
plt.plot(ts, y)
plt.plot(ts, y_lp)
plt.show()
```
