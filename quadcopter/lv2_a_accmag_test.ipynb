```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from imu.SO3 import SO3
from imu.SensorUtils import SensorUtils
from imu.RotationUtils import RotationUtils
np.set_printoptions(suppress=True, precision=2)
```

```python
data = pd.read_csv('data/imu_x.csv')

euler_gt = data[['euler_x', 'euler_y', 'euler_z']].values
acc = data[['accelerometer_x', 'accelerometer_y', 'accelerometer_z']].values
mag = data[['magnetometer_x', 'magnetometer_y', 'magnetometer_z']].values
```

```python
acc_si = SensorUtils.cvt_acc_si(acc)
N = acc_si.shape[0]


for i in range(N):
    euler = euler_gt[i] * np.pi / 180 + np.random.randn(3) * 0.01
    R_b2g = RotationUtils.eul2rotm(euler)
    acc_bdy = acc_si[i]
    #print(acc_bdy)
    acc_gnd = R_b2g @ acc_bdy
    print(acc_gnd)

    # error = acc_gnd - np.array([0, 0, -9.81])
    error_gnd = np.cross(acc_gnd, np.array([0, 0, -9.81]))
    error_bdy = R_b2g.T @ error_gnd
    dR = SO3.get_dR(error_bdy, 0.01)
    R_corr = R_b2g @ dR
    acc_gnd_corr = R_corr @ acc_bdy
    print(acc_gnd_corr)
```
