```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[0., 1], [0, 0]])
B = np.array([[0], [1]])
K = np.array([[2, 3]])
```

```python
dt = 0.01
time = np.arange(0, 10, dt)
#print(time)
N = len(time)
#print(N)
states = np.zeros((N,2))
states[0] = np.array([0, 0])
target = np.array([-100, 0])
g = 9.81
thrust = 0

for i in range(0, N-1):
    x = states[i]
    error = target - x
    #print(error.shape)
    u = K@error
    #print(u)
    x_dot = A @ x + B @ u
    #print(x_dot.shape)
    x_next = x + x_dot * dt
    states[i+1] = x_next

pos = states[:,0]
vel = states[:,1]

plt.figure()
plt.plot(time, pos)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')

plt.figure()
plt.plot(time, vel)
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')

```
