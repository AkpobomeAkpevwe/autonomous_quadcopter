# Observing Sensor Noise

```python
%matplotlib widget
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
np.set_printoptions(suppress=True, precision=2)
```

```python
#data = pd.read_csv('data/imu_noise_xyz.csv')
data = pd.read_csv('data/imu_xyz.csv')

ts = data['ts'].values
ts = ts - ts[0]
ts /=1000
euler_gt = data[['euler_x', 'euler_y', 'euler_z']].values
acc = data[['accelerometer_x', 'accelerometer_y', 'accelerometer_z']].values
gyro = data[['gyroscope_x', 'gyroscope_y', 'gyroscope_z']].values
mag = data[['magnetometer_x', 'magnetometer_y', 'magnetometer_z']].values


plt.figure()
plt.plot(ts, acc, label='acc_x')
plt.title('Accelerometer')

plt.figure()
plt.plot(ts, gyro, label='gyro_x')
plt.title('Gyroscope')

plt.figure()
plt.plot(ts, mag, label='mag_x')
plt.title('Magnetometer')
```
